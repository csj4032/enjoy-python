```python
def get_random_driver():
    browser_list = ["chrome", "firefox", "edge"]
    random_browser = random.choice(browser_list)
    if random_browser == "chrome":
        return webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
    elif random_browser == "firefox":
        return webdriver.Firefox(service=FirefoxService(GeckoDriverManager().install()))
    elif random_browser == "edge":
        return webdriver.Edge(service=EdgeService(EdgeChromiumDriverManager().install()))
```

```python
shuffled_articles = random.sample(articles, len(articles))
```

```python
for article in shuffled_articles:
    keyword = random.choice(article["keywords"])
    link =  article['link']
    print(f"{keyword}, {link}")
    driver = None
    try:
        driver = get_random_driver()
        driver.set_window_size(1280, 1600)
        driver.set_window_position(-1280, 0)
        driver.get("https://www.naver.com")
    
        search_box = WebDriverWait(driver, 1).until(EC.presence_of_element_located((By.ID, "query")))
        search_box.send_keys(keyword)
        search_box.send_keys(Keys.RETURN)
        time.sleep(random.uniform(1, 5))

        matched_element = None
        
        search_top_elements = driver.find_elements(By.CSS_SELECTOR, "a.link_tit")
        for element in search_top_elements:
            try:
                href = element.get_attribute("href")
                if href == link:
                    matched_element = element
                    print(f"{keyword} 검색 최상단 매치")
                    break
            except NoSuchElementException:
                continue

        if matched_element == None:
            search_elements = driver.find_elements(By.CSS_SELECTOR, "a.title_link")
            for element in search_elements:
                try:
                    href = element.get_attribute("href")
                    if href == link:
                        matched_element = element
                        print(f"{keyword} 검색 인기글 매치")
                        break
                except NoSuchElementException:
                    continue
        
        if matched_element == None:
            blog_tab = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.LINK_TEXT, "블로그")))
            blog_tab.click()
            time.sleep(random.uniform(2, 4))
            
            blog_elements = driver.find_elements(By.CSS_SELECTOR, "a.title_link")
    
            for element in blog_elements:
                try:
                    href = element.get_attribute("href")
                    if href == link:
                        matched_element = element
                        print(f"{keyword} 블로그 검색 매치")
                        break
                except NoSuchElementException:
                    continue

        if matched_element != None:
            WebDriverWait(driver, 5).until(EC.element_to_be_clickable(matched_element)).click()
            time.sleep(random.uniform(1, 10))
            WebDriverWait(driver, 5).until(EC.number_of_windows_to_be(2))
            driver.switch_to.window(driver.window_handles[1])
            main_iframe = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.ID, "mainFrame")))
            driver.switch_to.frame(main_iframe)
            scroll_height = random.uniform(4000, 8000)
            time.sleep(random.uniform(5, 10))
            driver.execute_script(f"window.scrollBy(0, {scroll_height});")
            time.sleep(random.uniform(5, 10))
            driver.switch_to.default_content()

    except Exception as e:
        print(f"{keyword} {link} 작업 중 오류 발생: {e}")
    finally:
        if driver:
            try:
                handles = driver.window_handles
                if len(handles) > 1:
                    driver.close()
                    driver.switch_to.window(handles[0])
                    driver.quit()
                else:
                    driver.quit()
            except Exception as e:
                print(f"세션 종료 중 오류 발생: {e}")
```
